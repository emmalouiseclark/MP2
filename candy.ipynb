```python
chocolate_candies = rawDF[rawDF['chocolate'] == 1]

print("Candies that have chocolate as an ingredient:")
print(chocolate_candies)
```

```python
fruity_candies = rawDF[rawDF['fruity'] == 1]

print("Candies that have fruit as an ingredient:")
print(fruity_candies)
```

```python
caramel_candies = rawDF[rawDF['caramel'] == 1]

print("Candies that have caramel as an ingredient:")
print(caramel_candies)
```

```python
nutty_candies = rawDF[rawDF['peanutyalmondy'] == 1]

print("Candies that have nuts as an ingredient:")
print(nutty_candies)
```

```python
nougat_candies = rawDF[rawDF['nougat'] == 1]

print("Candies that have nougat as an ingredient:")
print(nougat_candies)
```

```python
wafer_candies = rawDF[rawDF['crispedricewafer'] == 1]

print("Candies that have wafer as an ingredient:")
print(wafer_candies)
```

```python
hard_candies = rawDF[rawDF['hard'] == 1]

print("Candies that are hard:")
print(hard_candies)
```

```python
bar_candies = rawDF[rawDF['bar'] == 1]

print("Candies that are in bar form:")

print(bar_candies)
```

```python
candy_sums = rawDF[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar']].sum()

plt.figure(figsize=(10, 8))
plt.pie(candy_sums, labels=candy_sums.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Candies Based on Ingredients')
plt.axis('equal')  
plt.show()
```

```python
# Find the candy with the highest win percentage
most_popular_candy = rawDF.loc[rawDF['winpercent'].idxmax()]

print("The most popular candy is:")
print(most_popular_candy)

```

```python
# Find the candy with the lowest win percentage
least_popular_candy = rawDF.loc[rawDF['winpercent'].idxmin()]

print("The least popular candy is:")
print(least_popular_candy)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_candies = rawDF.sort_values('winpercent', ascending=False).head(10)
# Creating a blue gradient palette for win percentiles and reversing the order
palette = sns.color_palette("Blues", len(top_candies))[::-1]
sns.barplot(x='winpercent', y='competitorname', data=top_candies, palette=palette)
plt.xlabel('Win Percentage')
plt.ylabel('Candy')
plt.title('Top 10 Most Popular Candies')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

bottom_candies = rawDF.sort_values('winpercent').head(10)
# Creating a red gradient palette and reversing the order
palette = sns.color_palette("Reds", len(bottom_candies))[::-1]
sns.barplot(x='winpercent', y='competitorname', data=bottom_candies, palette=palette)
plt.xlabel('Win Percentage')
plt.ylabel('Candy')
plt.title('Top 10 Least Popular Candies')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_candies = rawDF.sort_values('pricepercent', ascending=False).head(10)
# Creating a blue gradient palette for price percentiles and reversing the order
palette = sns.color_palette("Greens", len(top_candies))[::-1]
sns.barplot(x='pricepercent', y='competitorname', data=top_candies, palette=palette)
plt.xlabel('Price Percentile')
plt.ylabel('Candy')
plt.title('Top 10 Candies with Highest Price Percentiles')
plt.show()
```
